```python
import pandas as pd
import PyPDF2 as Pypdf
import re
```

```python
def getPDFContent(path, first_page, last_page):
    content = []
    pdf = Pypdf.PdfFileReader(path)
    for i in range(first_page, last_page):
        content += pdf.getPage(i).extractText().splitlines()
    return content
```
